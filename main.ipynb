# Cancellation Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("hotel_booking.csv")

plt.style.use("ggplot")
```

### Basic Cancellation Analysis

```python
cancel_rate = df['is_canceled'].mean()
print("Overall Cancellation Rate: {:.2f}%".format(cancel_rate * 100))
```

### Cancellation Rate by hotel type

```python
cancel_by_hotel = df.groupby("hotel")["is_canceled"].mean()

sns.barplot(x=cancel_by_hotel.index, y=cancel_by_hotel.values)
plt.title("Cancellation Rate by Hotel Type")
plt.ylabel("Cancellation Rate")
plt.show()
```

### Relationship between special request and cancellation

```python
sns.barplot(x="total_of_special_requests", y="is_canceled", data=df)
plt.title("Special Requests vs Cancellation Rate")
plt.ylabel("Cancellation Rate")
plt.xlabel("Total Special Requests")
plt.show()
```

# ADR Analysis

### ADR by Hotel type

```python
adr_by_hotel = df.groupby("hotel")["adr"].mean().reset_index()

print("ADR by Hotel Type:")
print(adr_by_hotel)
```

### ADR by Customer Type

```python
adr_by_customer = df.groupby("customer_type")["adr"].mean().reset_index()

print("\nADR by Customer Type:")
print(adr_by_customer)
```

### ADR change by year

```python
adr_by_year = df.groupby("arrival_date_year")["adr"].mean().reset_index()

print("\nADR by Year:")
print(adr_by_year)
```

### Effect of cancellation on ADR

```python
adr_canceled = df[df["is_canceled"] == 1]["adr"].mean()
adr_not_canceled = df[df["is_canceled"] == 0]["adr"].mean()

print("ADR for Canceled Bookings:", round(adr_canceled, 2))
print("ADR for Not Canceled Bookings:", round(adr_not_canceled, 2))
print("Difference:", round(adr_canceled - adr_not_canceled, 2))

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(
    x=["Canceled", "Not Canceled"],
    y=[adr_canceled, adr_not_canceled],
    palette=["#d9534f", "#5cb85c"]
)
plt.title("Impact of Cancellation on ADR", fontsize=14)
plt.ylabel("Average Daily Rate (ADR)")
plt.show()
```

### ADR difference between hotel type

```python
city_adr = adr_by_hotel[adr_by_hotel["hotel"] == "City Hotel"]["adr"].values[0]
resort_adr = adr_by_hotel[adr_by_hotel["hotel"] == "Resort Hotel"]["adr"].values[0]

print(f"City ADR : {city_adr}")
print(f"Resort ADR : {resort_adr}")
adr_difference = round(city_adr - resort_adr, 2)

print(f"ADR Difference (City - Resort): {adr_difference}")
```

# Customer Segmentation

### Segmentation based on guest composition

```python
df["family_size"] = df["adults"] + df["children"] + df["babies"]
df["customer_group"] = df["family_size"].apply(
    lambda x: "Solo" if x == 1 else ("Couple" if x == 2 else "Family")
)

segment_counts = df["customer_group"].value_counts().reset_index()
segment_counts.columns = ["customer_group", "booking_count"] 

print(segment_counts)
print(segment_counts.columns) 

plt.figure(figsize=(8, 5))
sns.barplot(
    data=segment_counts,
    x="customer_group",
    y="booking_count",
    palette=["#4e79a7", "#f28e2b", "#76b7b2"]
)

plt.title("Customer Segmentation: Solo vs Couple vs Family", fontsize=14)
plt.xlabel("Customer Group")
plt.ylabel("Number of Bookings")
plt.tight_layout()
plt.show()
```

### Segmentatiion based on market type

```python
market_segment = df["market_segment"].value_counts().reset_index()
market_segment.columns = ["market_segment", "booking_count"]

print("\nMarket Segment Distribution:")
print(market_segment)

plt.figure(figsize=(10, 5))
sns.barplot(
    data=market_segment,
    x="market_segment",
    y="booking_count",
    palette="viridis"
)

plt.title("Market Segment Distribution", fontsize=14)
plt.xlabel("Market Segment")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Segmentation based on length of stay

```python
df["total_nights"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]

def stay_type(x):
    if x == 1:
        return "1-night stay"
    elif x <= 3:
        return "Short stay (2–3)"
    elif x <= 7:
        return "Medium stay (4–7)"
    else:
        return "Long stay (8+ nights)"

df["stay_segment"] = df["total_nights"].apply(stay_type)

stay_dist = df["stay_segment"].value_counts().reset_index()
stay_dist.columns = ["stay_segment", "booking_count"]

print("\nStay Duration Segmentation:")
print(stay_dist)

plt.figure(figsize=(9, 5))
sns.barplot(
    data=stay_dist,
    x="stay_segment",
    y="booking_count",
    palette=["#4e79a7", "#f28e2b", "#76b7b2", "#59a14f"]
)

plt.title("Stay Duration Distribution", fontsize=14)
plt.xlabel("Stay Duration Category")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

# Country Wise Analysis

### Top countries

```python
country_counts = df["country"].value_counts().reset_index()
country_counts.columns = ["country", "booking_count"]

print("\nTop 5 Countries by Booking Count:")
print(country_counts.head(5))

plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts["country"].head(5), y=country_counts["booking_count"].head(5))
plt.title("Top 5 Countries by Booking Count")
```

### Cancellation rate by country

```python
country_stats = df.groupby("country").agg(
    bookings=("is_canceled", "count"),
    cancellations=("is_canceled", "sum"),
    cancel_rate=("is_canceled", "mean")   # mean gives the rate [0..1]
).reset_index()

country_stats["cancel_rate_pct"] = country_stats["cancel_rate"] * 100

min_bookings = 30   # adjust as you like (e.g., 10, 30, 50)
country_stats_filtered = country_stats[country_stats["bookings"] >= min_bookings]

top5 = country_stats_filtered.sort_values("cancel_rate", ascending=False).head(5)

print("Top 5 Countries by Cancellation Rate (with counts):")
print(top5[["country", "bookings", "cancellations", "cancel_rate_pct"]].to_string(index=False))

plt.figure(figsize=(10,5))
sns.barplot(
    x="country",
    y="cancel_rate_pct",
    data=top5,
    order=top5["country"]
)
plt.ylabel("Cancellation Rate (%)")
plt.title(f"Top 5 Countries by Cancellation Rate (min bookings = {min_bookings})")
plt.ylim(0, 100)
for i, v in enumerate(top5["cancel_rate_pct"]):
    plt.text(i, v + 1, f"{v:.1f}%", ha="center")
plt.show()
```

### ADR by country

```python
adr_by_country = (
    df.groupby("country")["adr"]
      .mean()
      .reset_index()
      .sort_values("adr", ascending=False)
)

print("\nTop 5 Countries by ADR (Average Daily Rate):")
print(adr_by_country.head(5))

plt.figure(figsize=(12, 6))
sns.barplot(x=adr_by_country["country"].head(5), y=adr_by_country["adr"].head(5))
plt.title("Top 5 Countries by ADR (Average Daily Rate)")
```
